```python
import pandas as pd
import re

def extract_title(text):
    if not isinstance(text, str):
        return None
    
    # Split by newline and take the first meaningful line
    first_line = text.strip().split("\n")[0].strip()
    
    # Remove trailing words like "Summary"
    first_line = re.split(r'\bSummary\b', first_line, flags=re.IGNORECASE)[0].strip()
    
    return first_line

resume["title"] =resume["Resume_str"].apply(extract_title)
```

```python
def remove_title(row):
    resume = row["Resume_str"]
    title = row["title"]

    if not isinstance(resume, str) or not isinstance(title, str):
        return resume
    
    # Remove only the first occurrence of the title at the start
    cleaned = resume.strip()

    # If the resume starts with the title, remove it
    if cleaned.upper().startswith(title.upper()):
        cleaned = cleaned[len(title):].strip()

    return cleaned

resume["Resume_str"] = resume.apply(remove_title, axis=1)
```
